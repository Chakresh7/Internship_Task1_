```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path = '/content/Titanic-Dataset.csv'
df = pd.read_csv(path)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
sns.distplot(df['Age'])
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
sns.distplot(df['Age'])
```

```python
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
numeric_cols = ['Age', 'Fare']  # choose your numeric columns
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Fare'])
plt.title("Fare Boxplot")
plt.show()
```

```python
# Remove outliers for 'Age' using 3-sigma rule
df = df[(np.abs(df['Age'] - df['Age'].mean()) <= (3*df['Age'].std()))]

# Remove outliers for 'Fare' using IQR method
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['Fare'] >= lower_bound) & (df['Fare'] <= upper_bound)]
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Fare'])
plt.title("Fare Boxplot")
plt.show()
```

```python
df.to_csv("titanic_cleaned.csv", index=False)
```
